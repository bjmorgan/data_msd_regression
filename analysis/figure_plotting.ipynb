```python
import msd_analysis as msd
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

df_all = pd.read_csv('../data/msd_results_nsim16384_npart128.csv')
df_with = df_all[df_all.time_averaged]

# 1. Main variance comparison plot (OLS, WLS, GLS)
fig1 = msd.plot_all_methods(df_with)
plt.savefig('../figures/all_methods.pdf', dpi=300)
plt.show()

# 2. Time averaging comparison
fig2 = msd.plot_comparison_time_averaging(df_all)
plt.savefig('../figures/comparison_time_averaging.pdf', dpi=300)
plt.show()

# 3. OLS vs WLS-SqrtLag comparison
fig3 = msd.plot_ols_vs_wls_sqrtlag(df_all)
plt.savefig('../figures/ols_vs_wls_sqrtlag.pdf', dpi=300)
plt.show()

# 4. WLS weight scheme comparison
fig4 = msd.plot_wls_comparison(df_all)
plt.savefig('../figures/wls_comparison.pdf', dpi=300)
plt.show()

# 5. Comparison of Usler's "Variant A" and "Variant B"
# with WLS with correct weights (both time-averaged and non-time-averaged)
fig5 = msd.plot_comprehensive_wls_comparison(df_all)
plt.savefig('../figures/comprehensive_wls_comparison.pdf', dpi=300)
plt.show()

# 6. Comparing OLS, WLS, GLS for Δt_max = t_sim with time-averaging
fig6 = msd.plot_methods_single_panel(df_all)
plt.savefig('../figures/methods_comparison_single_panel.pdf', dpi=300)
plt.ylim(5e-6,3.3e-2)
plt.show()
```
